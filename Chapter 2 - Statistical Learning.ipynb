```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Problem 7

```python
# Create dataframe for given training data
df7 = pd.DataFrame({'x1': [0, 2, 0, 0, -1, 1], 
                    'x2': [3, 0, 1, 1, 0, 1], 
                    'x3': [0, 0, 3, 2, 1, 1], 
                    'y' : ['Red', 'Red', 'Red', 'Green', 'Green', 'Red']})

# test point x1 = x2 = x3 = 0
x = [0, 0, 0]

# Problem 7 a
# Compute euclidean distance between test point and all points in training data
df7['euclidean_dist'] = np.sqrt( (df7.x1-x[0])**2 + (df7.x2-x[1])**2 + (df7.x3-x[2])**2)

#Sort values in ascending order based on Eucledian Distance
df7.sort_values(by = 'euclidean_dist',inplace = True)
df7.reset_index(drop = True, inplace= True)
df7
```

```python
#Problem 7 b , 7c
#Set value of k 
k = 3 

#Prediction for the given k value
knn = df7[0:k]
knn = knn.groupby('y')['y'].agg({'count'}).reset_index()
knn.sort_values(by ='count',ascending = False,inplace = True)
knn.iloc[0]['y'] # If probability is same , y which comes first will be predicted
```

```python
df8 = pd.read_csv('http://www-bcf.usc.edu/~gareth/ISL/College.csv',index_col=0)
df8.describe()
```

```python
# 8c ii
sns.pairplot(df8.iloc[0:,0:10]);
```

```python
#8c iv
df8['Elite'] = df8['Top10perc'].apply(lambda x : 'Yes' if x > 50 else 'No')
df8.describe(include = 'all')
```

```python
sns.boxplot('Elite','Outstate', data = df8)
```

```python
#8c v
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,8))

df8['Accept'].hist(ax=axes[0,0] ,bins = 10);axes[0,0].set_title('Accept')
df8['Accept'].hist(ax=axes[0,1] , bins = 20);axes[0,1].set_title('Accept')
df8['Accept'].hist(ax=axes[1,0] , bins = 30);axes[1,0].set_title('Accept')
df8['Accept'].hist(ax=axes[1,1] , bins = 50);axes[1,1].set_title('Accept')
```

```python
#8c vi
sns.regplot(df8['S.F.Ratio'], df8['Grad.Rate'], order=1, ci=None, scatter_kws={'color':'g'})
```
